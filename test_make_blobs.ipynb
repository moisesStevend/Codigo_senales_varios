```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the function to generate the blobs
def generate_blobs(num_blobs):
    blobs = []
    for i in range(num_blobs):
        x = np.random.randint(-40, 41)
        y = np.random.randint(-40, 41)
        r = np.random.randint(5, 16)
        blobs.append((x, y, r))
    return blobs

# Define the function to create the blob image
def create_blob_image(x, y, r):
    x_vals, y_vals = np.meshgrid(np.arange(-50, 51), np.arange(-50, 51))
    circle = (x_vals - x)**2 + (y_vals - y)**2 <= r**2
    return circle

# Define the function to create the scatter plot of the blobs
def create_scatter_plot(blobs):
    fig, ax = plt.subplots(figsize=(6, 6))
    for blob in blobs:
        ax.scatter(blob[0], blob[1], color='red', alpha=0.5)
    ax.set_xlim((-50, 50))
    ax.set_ylim((-50, 50))
    ax.set_aspect('equal')
    plt.show()

# Define the function to fit the polynomial model
def fit_polynomial_model(blobs):
    x = [blob[0] for blob in blobs]
    y = [blob[1] for blob in blobs]
    popt, pcov = curve_fit(lambda t,a,b,c: a*t**2 + b*t + c, x, y)
    return popt

# Define the widgets for the Blobs tool
num_blobs_slider = widgets.IntSlider(value=3, min=1, max=10, step=1, description='Number of blobs:')
generate_button = widgets.Button(description='Generate')
fit_button = widgets.Button(description='Fit model')

# Define the event handler for the Generate button
def on_generate_button_click(b):
    global blobs
    with output:
        clear_output(wait=True)
        num_blobs = num_blobs_slider.value
        blobs = generate_blobs(num_blobs)
        create_scatter_plot(blobs)

# Define the event handler for the Fit model button
def on_fit_button_click(b):
    with output:
        clear_output(wait=True)
        popt = fit_polynomial_model(blobs)
        x_vals = np.linspace(-50, 50, 100)
        y_vals = popt[0]*x_vals**2 + popt[1]*x_vals + popt[2]
        fig, ax = plt.subplots(figsize=(6, 6))
        for blob in blobs:
            ax.imshow(create_blob_image(*blob), extent=[blob[0]-blob[2], blob[0]+blob[2], blob[1]-blob[2], blob[1]+blob[2]])
        ax.plot(x_vals, y_vals, color='blue')
        ax.set_xlim((-50, 50))
        ax.set_ylim((-50, 50))
        ax.set_aspect('equal')
        plt.show()

# Attach the event handlers to the buttons
generate_button.on_click(on_generate_button_click)
fit_button.on_click(on_fit_button_click)

# Create the layout for the Blobs tool
ui = widgets.VBox([num_blobs_slider, generate_button, fit_button])
```
